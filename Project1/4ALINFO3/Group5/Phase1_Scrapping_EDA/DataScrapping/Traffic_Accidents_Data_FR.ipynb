```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
import time

driver = webdriver.Chrome()
driver.get("https://www.data.gouv.fr/fr/datasets/bases-de-donnees-annuelles-des-accidents-corporels-de-la-circulation-routiere-annees-de-2005-a-2022/")
current_page = 1
target_page = 15  

while True:
    try:
        # Wait for the article elements to be present
        article_elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//*[@id="resources-panel"]/div/div/section[1]/div[3]/article'))
        )

        # Iterate over each article and extract information
        for i, article_element in enumerate(article_elements, start=1):
            try:
                file_name_element = article_element.find_element(By.CSS_SELECTOR, 'h4')
                file_name = file_name_element.text
                download_buttons = article_element.find_elements(By.XPATH, './/p/a[contains(@class, "matomo_download")]')

                # Iterate through all download buttons for the current element
                for download_button in download_buttons:
                    file_url = download_button.get_attribute('href')
                    
                    # Open the download link directly
                    driver.get(file_url)
                    time.sleep(5)  # Adjust the sleep time as needed

                    print(f"Page {current_page} - File {i} - Name: {file_name} - Done")

            except StaleElementReferenceException:
                pass
        if current_page == target_page:
                    print(f"Reached the target page {target_page} ")
                    break
        # Check if the "Next" button is enabled
        try:
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CLASS_NAME, 'fr-pagination__link--next'))
            )
            if "fr-pagination__link--disabled" in next_button.get_attribute("class"):
                print("Reached the last page. Stopping.")
                break  # Break the loop if the "Next" button is disabled (last page)

            driver.execute_script("arguments[0].scrollIntoView();", next_button)
            next_button.click()

            # Wait for the new page to load and establish the context again
            WebDriverWait(driver, 10).until(
                EC.staleness_of(article_elements[0])
            )

            current_page += 1

        except NoSuchElementException:
            break  # Break the loop if the "Next" button is not found
        except ElementNotInteractableException:
            WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.CLASS_NAME, 'fr-pagination__link--next'))
            )
            continue

    except StaleElementReferenceException:
        pass
```
