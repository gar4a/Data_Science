```python
import pandas as pd
```

```python
file_path = 'dataset.csv'
df = pd.read_csv(file_path, encoding='latin1')
df
```

```python
#Remove 0 latitude and long value rows
latitude_column = 'Latitude'
longitude_column = 'Longitude'
df = df[df[latitude_column] != 0]
df = df[df[longitude_column] != 0]
df
```

```python
#change null values to unkwonw in primary reason and reported location
factor_name = 'Primary Factor'
df[factor_name].fillna('unknown', inplace=True)
df
```

```python
def get_hours(nbr):
    if isinstance(nbr, pd.Series):
        result = nbr.apply(get_hours)
        return result
    
    if pd.isna(nbr) or nbr == "":
        return "00"
    
    if (nbr == "nan"):
        return "00"
    
    value = nbr.replace('.', '')
    value = value.replace('0', '')
    value = "0" if value == '' else value
    return "0" + value if len(value) == 1 else value
```

```python
#add new column full date
df['Year'] = df['Year'].astype(str)
df['Month'] = df['Month'].astype(str)
df['Day'] = df['Day'].astype(str)
df['Hour'] = df['Hour'].astype(str)
df['Full_Date'] = pd.to_datetime(df['Year'] + '-' + df['Month'] + '-' + df['Day'] + ' ' + get_hours(df['Hour'])  + ':00')
df
```

```python
#remove blanks rows in collision type
df = df[df['Collision Type'].notna()]
df
```

```python
#change null values in weekend to default weekend and change to yes or no
weekend = 'Weekend?'
df[weekend].fillna('Weekend', inplace=True)
day_mapping = {'Weekend': 'yes', 'Weekday': 'no'}
df['Weekend?'] = df['Weekend?'].map(day_mapping)
df
```

```python
#add new column number of vehicules involved
collision_mapping = {'1-Car': "1", '2-Car': "2", '3+ Cars': "3+", "Bus" : "2", "Cyclist" : "2",
                     "Moped/Motorcycle" : "2", "Pedestrian" : "1"}

df['nbr_of_vehicles'] = df['Collision Type'].map(collision_mapping)
df
```

```python
df.to_csv('cleaned.csv', index=False)
```
